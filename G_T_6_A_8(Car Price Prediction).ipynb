How Much is Your Car Worth?

Data about the retail price of 2005 General Motors cars can be found in car_data.csv.

The columns are:

1)Price: suggested retail price of the used 2005 GM car in excellent condition.

2)Mileage: number of miles the car has been driven

3)Make: manufacturer of the car such as Saturn, Pontiac, and Chevrolet

4)Model: specific models for each car manufacturer such as Ion, Vibe, Cavalier

5)Trim (of car): specific type of car model such as SE Sedan 4D, Quad Coupe 2D
Type: body type such as sedan, coupe, etc.

6)Cylinder: number of cylinders in the engine

7)Liter: a more specific measure of engine size

8)Doors: number of doors

9)Cruise: indicator variable representing whether the car has cruise control (1 = cruise)

10)Sound: indicator variable representing whether the car has upgraded speakers (1 = upgraded)

11)Leather: indicator variable representing whether the car has leather seats (1 = leather)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
data=pd.read_csv("/content/car_data.csv")
```

```python
data.head()
```

```python
data.corr()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
X=data.iloc[:,1] #X=mileage(independent variable)
y=data.iloc[:,0] #y=price(dependent variable)
```

```python
X=np.array(X).reshape((-1,1))
y=np.array(y).reshape((-1,1))
```

```python
X
```

```python
y
```

```python
#model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,y)
```

```python
print(model.coef_)#y=mx+c ->>m
```

```python
print(model.intercept_) #y=mx+c->>c
```

```python
model.score(X,y)
```

```python
print("Linear Regression Equation for Mileage vs price is",
      "y=",float(model.coef_),"x +",float(model.intercept_))
```

```python
pred=model.predict(X)
pred
from sklearn.metrics import r2_score
print("The R2 score of mileage column is :",r2_score(y,pred))
```

```python
print(data['Make'].unique())
print(data['Model'].unique())
print(data['Trim'].unique())
print(data['Type'].unique())
```

```python
encoded_data = pd.get_dummies(data, columns = ['Make','Model','Trim','Type'])
```

```python
encoded_data.head()
```

```python
X2 = encoded_data.iloc[:,1:99]
y2 = encoded_data.iloc[:,0]

X2 = np.array(X2).reshape((-1,97))
y2 = np.array(y2).reshape((-1,1))
```

```python
s = []
for i in range(1,98):
  pca2 = PCA(n_components=i,random_state=0)
  pca2.fit(X2)
  x2=pca2.transform(X2)

  model3 = LinearRegression()
  model3.fit(x2,y2)

  pred3 = model3.predict(x2)

  s.append(r2_score(y2, pred3))
  print("With",i,"columns:",s[i-1])
```

```python
for i in s:
  if i>=0.992:
      print("With",i,"we have:",s.index(i)+1,"columns out of 98")
```
